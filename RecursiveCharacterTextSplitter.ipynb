```python
### Import Packages
from langchain_community.document_loaders import UnstructuredWordDocumentLoader, PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

__Global Variables__

```python
## Global Variables
document_name = 'responsible_finance'
```

__Document Loader__

```python
## Loading documents -- Each page of the ppdf is considered a document.
documents = PyPDFLoader(document_name+".pdf").load()
```

__Text Splitter__

```python
## Recursive Splitter 
chunk_size = 500
chunk_overlap = 50

splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size,\
                                          chunk_overlap=chunk_overlap,\
                                          separators='\n \n')
```

```python
## Create Chunks on the document
chunks = splitter.split_documents(documents)
```

```python
## Display Chunks
for i, chunk in enumerate(chunks):
    print('-------------Chunk ID#{}-----------'.format(i+1))
    print(chunk.page_content[:chunk_size])
```
