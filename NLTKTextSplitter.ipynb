```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import NLTKTextSplitter
```

```python
### Essesntial for using NLTKTextSplitter
import nltk
nltk.download('punkt_tab')
```

__Load Text Document Using TextLoader__
TextLoader returns Document Type which maintains relationship between segment/pages/sections via. key value pairs. It also stores metadata and actual content is available in tag: page_content

```python
documents = TextLoader('responsible_finance.txt').load()
```

__Create Splitter Opbject__

```python
text_splitter = NLTKTextSplitter(chunk_size=300, ## number of characters per chunk
                                chunk_overlap=50 ## this preserves context
                                )
```

__Split Document and Display__

```python
chunks = text_splitter.split_documents(documents)
```

```python
for i, doc in enumerate(chunks):
    print('\n\n=== Chunk {} ===='.format(i+1))
    print(doc.page_content[:300])
```
