```python
import pandas as pd
from langchain_community.document_loaders import PyPDFLoader, TextLoader, UnstructuredWordDocumentLoader
```

```python
### init a blank list
all_doc_list = []
document_name = 'responsible_finance'
```

__PDF Document Load__

```python
pdf_doc = PyPDFLoader(document_name+".pdf").load() ## The number of documents should be equal to number of pages in the pdf.
```

```python
## Appending to List
all_doc_list.extend(pdf_doc)
```

__Text Document Loading__

```python
text_doc = TextLoader(document_name+".txt").load() ## The number of documents should be equal to number of paragraphs in the text file.
```

```python
## Appending to List
all_doc_list.extend(text_doc)
```

__Word Document Loading__

```python
word_doc = UnstructuredWordDocumentLoader(document_name+".docx").load() ## The number of documents should be equal to number of sections in the text file.
```

```python
## Appending to List
all_doc_list.extend(word_doc)
```

__Printing all documents__

```python
for i, doc in enumerate(all_doc_list[:5]):
    print('Document {}:: {}\n\n'.format(i, doc))
```
