```python
# !pip install langchain_community
# !pip install tiktoken
from langchain_text_splitters import TokenTextSplitter
```

TokenTextSplitter split the text into chunks based on token size.

__Global Variables__

```python
texts = """Retrieval-augmented generation (RAG) is a technique that enables large language models (LLMs) to retrieve and incorporate new information from external data sources.[1] With RAG, LLMs first refer to a specified set of documents, then respond to user queries. These documents supplement information from the LLM's pre-existing training data.[2] This allows LLMs to use domain-specific and/or updated information that is not available in the training data.[2] For example, this helps LLM-based chatbots access internal company data or generate responses based on authoritative sources."""
```

__Text Splitter__

```python
splitter = TokenTextSplitter(chunk_size=20, chunk_overlap=5)
```

```python
splits = splitter.split_text(texts)
```

```python
splits
```

```python
for i, split in enumerate(splits):
    print('_____________{}__________________'.format(i+1))
    print('\n\n', split)
```
