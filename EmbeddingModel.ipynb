## Text EMbedding Using HuggingFaceEmbeddings Wrapper from Langchain

Embedding is a process of converting text to multi-dimentional vectors that captures their **semantic meanings**.

```python
# !pip install langchain sentence-transformers
# !pip install langchain-huggingface
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

__Sample Text__

```python
text = """Retrieval-Augmented Generation (RAG) for Wikipedia involves building AI systems that retrieve relevant, real-time data from 
Wikipedia to ground Large Language Model (LLM) responses, 
significantly reducing hallucinations and providing verifiable, up-to-date answers. 
Key implementations utilize tools like LangChain, FAISS, Ollama, and vector databases to index Wikipedia content, 
enabling semantic search over encyclopedic data"""
```

__Embedding Model__

```python
emb_model_name = "all-MiniLM-L6-v2"
```

```python
embedding_model  = HuggingFaceEmbeddings(model=emb_model_name)
```

__Embedded Text__

```python
embeddings = embedding_model.embed_documents(text)
```

__Displaying Embeddings__

```python
for i, vector in enumerate(embeddings):
    print('\n\n=== Embeddings {} (first 5 dimension) ===='.format(i))
    print(vector[:6])
```
