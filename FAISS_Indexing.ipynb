```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

###### 1. Sample texts to embed

```python
texts = [
    "Retreieval Augmented Generation (RAG) is an architecture that combines the ability of large language models (LLMs) with a retrieval system to enhance the factual accuracy.",
     "Traditional generattive models rely solely on internal paraameters for producing responses, which limits their ability to provide up-to-date or domain-specific knoledge.",
     "Rag mitigates this by augmenting the generation process with real-time retrival from external knowledge sources.",
     "Tradition generative models are now mostly replaced or augmented by deep learning-based transformer models, which offer greater accuracy, coherence, and scalability.",
     "Data center plays important role in today's digitalisation era."
        ]
```

###### 2. Load embedding model

```python
embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-V2")
```

###### 3. Create FAISSS Index from the embeddings

```python
vector_store = FAISS.from_texts(texts, embedding=embedding_model)
```

###### 4. Similarity Search

```python
user_query = "What is RAG?"
```

```python
## Search result , top 3
results = vector_store.similarity_search(user_query, k=3)
```

###### 5. Output of the search

```python
print("Query: ", user_query)
for i, doc in enumerate(results):
    print('Result {}: {}'.format(i+1, doc.page_content))
```
